```python
def makefeatfile(ascat, centromere, gccontent):

    feature_df = pd.DataFrame({
              'Sample': ascat['ID'],
                 'Chr': ascat['Chr'],
                  'CN': ascat['cn'], 
             'SegSize': ascat['End'] - ascat['Start'],
           'Dist2Cent': feature.getDist2Centromere(ascat, centromere),
              'SegVal': feature.getSegVal(ascat),
                 'LOH': feature.getLOH(ascat),
          'SizeDipSeg': feature.getSizeofDiploidSeg(ascat),
                'CpCN': feature.getCP(ascat),
           'Dist2nCNV': feature.getDist2CNV(ascat),
              'GCcSeg': feature.getGCcontent(ascat, gccontent)
     
    })

    feature_df = feature_df.groupby(['Sample', 'Chr'], as_index=False, sort=False).mean()
 
    return feature_df

df = makefeatfile(ascat, centAnno, gc)

df
```

```python
def no_repeats(ascat, line1):
    reps = [0] * len(ascat)
    line1 = line1[line1['repFamily']=='L1'].reset_index()
    line1['Chr'] = line1['genoName'].str.split('_').str[0].str.extract('(\d+)', expand=False)
    
    for i in range(0, len(ascat)):
        count = 0
        for j in range(0, len(line1)):
            if int(line1['Chr'][j]) == ascat['Chr'][i] and line1['genoStart'][j] >= ascat['Start'][i] and line1['genoStart'][j] < ascat['End'][i]:
                count += 1
            elif int(line1['Chr'][j]) == ascat['Chr'][i] and line1['genoStart'][j] < ascat['Start'][i] and line1['genoEnd'][j] > ascat['Start'][i]:
                count += 1
            else:
                continue
        reps[i] = count
    
    
    return reps
```
