```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_table('../data/filteredAscatRaw.txt', dtype=str)
```

```python
def sample(df):
    cancertypes = np.unique(df['cancer_type'])
    samples = df.groupby(['ID'])
    
    sampled_df = []
    for i in range(len(cancertypes)):
        subset = df[df.cancer_type == cancertypes[i]].sample(n=100, random_state=17).values.tolist()
        for j in subset:
            sampled_df.append(j)
    
    subset_df = pd.DataFrame(sampled_df, columns=df.columns)
    final_df = df[df['ID'].isin(subset_df['ID'])]
    
    return final_df
```

```python
new_df = sample(file)
new_df
```
